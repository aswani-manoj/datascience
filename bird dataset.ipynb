```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Birdddataset.csv")
plt.figure(figsize=(10,3))
x=data["Number of bird species"]
plt.hist(x,bins=30,color="green")
plt.xlabel("Distance from transect (m)")
plt.ylabel("count")
plt.show()
```
